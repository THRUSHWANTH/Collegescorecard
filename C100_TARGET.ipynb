```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('modified_file.csv',low_memory=False)
```

```python
df.head()
```

```python
df.describe()
```

```python
# Summary statistics for the continuous target variable
target_description = df['C100_4_POOLED_SUPP'].describe()
print(target_description)
print(df['C100_4_POOLED_SUPP'][:10])
```

```python

# Replace 'PrivacySuppressed' with NaN
df['C100_4_POOLED_SUPP'] = pd.to_numeric(df['C100_4_POOLED_SUPP'], errors='coerce')

```

```python
missing_count = df['C100_4_POOLED_SUPP'].isna().sum()
total_count = len(df)
missing_prop = missing_count / total_count

print(f"Missing Values: {missing_count}")
print(f"Proportion of Missing Values: {missing_prop:.2%}")

```

```python
df = df.dropna(subset=['C100_4_POOLED_SUPP'])
```

```python
missing_after_removal = df['C100_4_POOLED_SUPP'].isna().sum()
print(f"Missing values after removal: {missing_after_removal}")
```

```python
print(df['C100_4_POOLED_SUPP'].describe())
print(df['C100_4_POOLED_SUPP'][:10])
```

```python
sns.histplot(df['C100_4_POOLED_SUPP'], kde=True)
plt.title("Distribution of C100_4_POOLED_SUPP after Removal of Missing Values")
plt.show()
```

```python
sns.boxplot(x=df['C100_4_POOLED_SUPP'])
plt.title("Boxplot of C100_4_POOLED_SUPP after Removal of Missing Values")
plt.show()
```

```python
Q1 = df['C100_4_POOLED_SUPP'].quantile(0.25)
Q3 = df['C100_4_POOLED_SUPP'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df['C100_4_POOLED_SUPP'] < lower_bound) | (df['C100_4_POOLED_SUPP'] > upper_bound)]
print(f"Number of Outliers: {len(outliers)}")
```

```python
skewness = df['C100_4_POOLED_SUPP'].skew()
print(f"Skewness: {skewness}")
```

```python
# Moving target variable to the last 
cols = list(df.columns)
cols_to_move = ['COMP_ORIG_YR4_RT', 'C100_4_POOLED_SUPP']
cols = [col for col in cols if col not in cols_to_move] + cols_to_move

df = df[cols]
df = df.reset_index(drop=True)

```

```python
df.to_csv('cleaned_C100.csv',index=False)
```
